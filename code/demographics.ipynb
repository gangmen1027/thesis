```python
import pandas as pd
import numpy as np
```

```python
CA = pd.read_csv('mimic CA demo.csv')
CA
```

```python
CA.describe()
```

```python
import scipy.stats

# 使用Shapiro檢查常態分布
def shapiro_test(data):
    stat, p = scipy.stats.shapiro(data)
    return p
```

```python
# 將每個features的NA值去除
p_hr = shapiro_test(CA['heart_rate'].dropna().tolist())
p_temp = shapiro_test(CA['temperature'].dropna().tolist())
p_sbp = shapiro_test(CA['sbp_ni'].dropna().tolist())
p_dbp = shapiro_test(CA['dbp_ni'].dropna().tolist())
p_rr = shapiro_test(CA['resp_rate'].dropna().tolist())
p_spo2 = shapiro_test(CA['spo2'].dropna().tolist())

print('CA：'+ str(p_hr))
print('BT：'+ str(p_temp))
print('SBP：'+ str(p_sbp))
print('DBP：'+ str(p_dbp))
print('RR：'+ str(p_rr))
print('spo2：'+ str(p_spo2))
```

```python
# 選擇每位患者
CA1 = CA.drop_duplicates(subset='subject_id', keep='first')
CA1
```

```python
# 計算患者性別數量和比例
print(CA1['gender'].value_counts(normalize=True))
print(' ')
print('男性數量')
print((CA1['gender'] == 'M').sum())
```

```python
# 計算患者年齡平均
print(CA1['anchor_age'].mean())
print(CA1['anchor_age'].std())
```

```python
from scipy.stats import levene

# CA和non-CA相比：變異數是否相同

# 年紀(不用比，因為是配對的)
statistic, p_value = levene(CA1['anchor_age'], noCA1['anchor_age'])
# 六個vital signs
hstatistic, hp_value = levene(CA['heart_rate'].dropna().tolist(), noCA['heart_rate'].dropna().tolist())
sstatistic, sp_value = levene(CA['sbp_ni'].dropna().tolist(), noCA['sbp_ni'].dropna().tolist())
dstatistic, dp_value = levene(CA['dbp_ni'].dropna().tolist(), noCA['dbp_ni'].dropna().tolist())
rstatistic, rp_value = levene(CA['resp_rate'].dropna().tolist(), noCA['resp_rate'].dropna().tolist())
tstatistic, tp_value = levene(CA['temperature'].dropna().tolist(), noCA['temperature'].dropna().tolist())
ostatistic, op_value = levene(CA['spo2'].dropna().tolist(), noCA['spo2'].dropna().tolist())

print('統計量：%.2f'%statistic)
print('p-value:%.5f'%p_value)
print('-'*30)

print('統計量：%.2f'%hstatistic)
print('p-value:%.5f'%hp_value)
print('-'*30)
print('統計量：%.2f'%sstatistic)
print('p-value:%.5f'%sp_value)
print('-'*30)
print('統計量：%.2f'%dstatistic)
print('p-value:%.5f'%dp_value)
print('-'*30)
print('統計量：%.2f'%rstatistic)
print('p-value:%.5f'%rp_value)
print('-'*30)
print('統計量：%.2f'%tstatistic)
print('p-value:%.5f'%tp_value)
print('-'*30)
print('統計量：%.2f'%ostatistic)
print('p-value:%.5f'%op_value)
print('-'*30)
```

```python
from scipy.stats import ttest_ind

# CA和non-CA相比：獨立t test

print(ttest_ind(CA1['anchor_age'], noCA1['anchor_age'], equal_var=False, alternative='two-sided'))
print('-'*30)

print(ttest_ind(CA['heart_rate'].dropna().tolist(), noCA['heart_rate'].dropna().tolist(), equal_var=False, alternative='two-sided'))
print('-'*30)
print(ttest_ind(CA['sbp_ni'].dropna().tolist(), noCA['sbp_ni'].dropna().tolist(), equal_var=False, alternative='two-sided'))
print('-'*30)
print(ttest_ind(CA['dbp_ni'].dropna().tolist(), noCA['dbp_ni'].dropna().tolist(), equal_var=True, alternative='two-sided'))
print('-'*30)
print(ttest_ind(CA['resp_rate'].dropna().tolist(), noCA['resp_rate'].dropna().tolist(), equal_var=False, alternative='two-sided'))
print('-'*30)
print(ttest_ind(CA['temperature'].dropna().tolist(), noCA['temperature'].dropna().tolist(), equal_var=False, alternative='two-sided'))
print('-'*30)
print(ttest_ind(CA['spo2'].dropna().tolist(), noCA['spo2'].dropna().tolist(), equal_var=False, alternative='two-sided'))
```
